```python
import pandas as pd
from datetime import datetime
```

```python
sinistri = pd.read_csv('sinistri.csv', delimiter=';')
sinistri['targa'] = sinistri['COD_TARGA_TELAIO']
del sinistri['Unnamed: 13']
sinistri.head()
```

```python
targhe = pd.read_csv('targa2voucher.csv', delimiter=';')
targhe.head()
```

```python
df = pd.merge(sinistri, targhe, on='targa', how='left')
df.head()
```

```python
len(df)
```

```python
df['data'] = pd.to_datetime(df['DAT_AVVENIM_DANNO'], format='%d/%m/%Y %H:%M')
df.head()
```

```python
df = df[df['data']<=datetime.strptime('2018-10-31', '%Y-%m-%d')]
df = df[df['data']>=datetime.strptime('2018-02-12', '%Y-%m-%d')]
len(df)
```

```python
df.head()
```

```python
dfd = df.dropna(subset=['contractid','DAT_AVVENIM_DANNO'])
len(dfd)
```

```python
luoghi = dfd.dropna(subset=['DES_LUOGO_DANNO'])['DES_LUOGO_DANNO']
l = list(luoghi)
l = [luogo.strip().strip(',') for luogo in l if luogo.strip().strip(',')\
     not in ['VIA', 'MILANO', 'IGNOTO', 'IGNOTA','LUOGO NON INDICATO', 'X']]
print(len(l))
l
```
